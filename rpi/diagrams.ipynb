```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('test1_monitoring.txt', sep=', ')

df['CPU-Usage'] = df['CPU-Usage'].str.rstrip('%').astype('float')
df['Memory'] = df['Memory'].str.rstrip('%').astype('float')

plt.figure(figsize=(10,6))

plt.plot(df['CPU-Usage'], label='CPU-Usage (%)')

plt.plot(df['Memory'], label='Memory (%)')

plt.title('Single Signings')
plt.xlabel('Timestamps')
plt.ylabel('Workload in percentage')
plt.yticks([0, 5, 10, 15, 20, 25, 30])
plt.legend()
plt.show()
```
